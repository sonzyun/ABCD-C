```python
import pandas as pd

# ë‹¤ì–‘í•œ ì¸ì½”ë”©ì„ ì‹œë„í•´ë³´ëŠ” í•¨ìˆ˜
def read_csv_with_encoding(file_path):
    encodings = ['utf-8', 'cp949', 'euc-kr', 'utf-8-sig', 'latin1']
    
    for encoding in encodings:
        try:
            print(f"{encoding} ì¸ì½”ë”©ìœ¼ë¡œ ì‹œë„ ì¤‘...")
            df = pd.read_csv(file_path, encoding=encoding)
            print(f"âœ“ {encoding} ì¸ì½”ë”©ìœ¼ë¡œ ì„±ê³µ!")
            return df
        except Exception as e:
            print(f"âœ— {encoding} ì‹¤íŒ¨: {str(e)}")
            continue
    
    print("ëª¨ë“  ì¸ì½”ë”© ì‹¤íŒ¨. íŒŒì¼ì„ í™•ì¸í•´ì£¼ì„¸ìš”.")
    return None

# CSV íŒŒì¼ ì½ê¸°
df = read_csv_with_encoding('/Users/sonzyun/Desktop/VScode/ABCD_C/25.09.09/data(Sheet1).csv')

if df is not None:
    # ê¸°ë³¸ ë°ì´í„° ì •ë³´
    print("\n" + "="*50)
    print("ë°ì´í„° í¬ê¸°:", df.shape)
    print("\nì²« 5í–‰:")
    print(df.head())
    print("\nì»¬ëŸ¼ëª…:")
    print(df.columns.tolist())
    
    # ì²« ë²ˆì§¸ ì»¬ëŸ¼ì„ ì´ë¦„ìœ¼ë¡œ, ë‚˜ë¨¸ì§€ë¥¼ ìˆ«ìí˜• ì»¬ëŸ¼ìœ¼ë¡œ ë¶„ë¦¬
    name_col = df.columns[0]  # ì²« ë²ˆì§¸ ì»¬ëŸ¼ (ì´ë¦„)
    numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
    
    print(f"\nì´ë¦„ ì»¬ëŸ¼: {name_col}")
    print(f"ìˆ«ìí˜• ì»¬ëŸ¼: {numeric_cols}")
    
    if len(numeric_cols) > 0:
        # 1. ê° ì‚¬ëŒ(ë ˆì½”ë“œ)ë³„ í‰ê·  ê³„ì‚°
        print("\n" + "="*50)
        print("ğŸ‘¥ ê° ì‚¬ëŒë³„ í‰ê· ì ìˆ˜:")
        print("-" * 40)
        
        df_numeric = df[numeric_cols]
        person_averages = df_numeric.mean(axis=1)  # ê° í–‰(ì‚¬ëŒ)ì˜ í‰ê· 
        
        # ì´ë¦„ê³¼ í‰ê· ì„ í•¨ê»˜ ì¶œë ¥
        for i, (name, avg) in enumerate(zip(df[name_col], person_averages)):
            print(f"{name}: {avg:.2f}")
        
        # 2. ì‚¬ëŒë³„ í‰ê· ì„ ë°ì´í„°í”„ë ˆì„ì— ì¶”ê°€
        df['ê°œì¸í‰ê· '] = person_averages
        
        # 3. ì»¬ëŸ¼ë³„ í‰ê· ê°’ (ê³¼ëª©ë³„ í‰ê· )
        print("\n" + "="*50)
        print("ğŸ“Š ê³¼ëª©ë³„ ì „ì²´ í‰ê· :")
        print("-" * 30)
        for col in numeric_cols:
            mean_val = df[col].mean()
            print(f"{col}: {mean_val:.2f}")
        
        # 4. ê°œì¸ë³„ ìƒì„¸ ì ìˆ˜ì™€ í‰ê·  í…Œì´ë¸”
        print("\n" + "="*50)
        print("ğŸ“‹ ê°œì¸ë³„ ìƒì„¸ ì ìˆ˜í‘œ:")
        print("-" * 50)
        result_df = df[[name_col] + numeric_cols + ['ê°œì¸í‰ê· ']].copy()
        print(result_df.to_string(index=False, float_format='%.2f'))
        
        # 5. ê°œì¸ í‰ê· ì˜ í†µê³„
        print("\n" + "="*50)
        print("ğŸ“ˆ ê°œì¸ í‰ê·  ì ìˆ˜ í†µê³„:")
        print("-" * 30)
        print(f"ì „ì²´ ê°œì¸í‰ê· ì˜ í‰ê· : {person_averages.mean():.2f}")
        print(f"ê°€ì¥ ë†’ì€ ê°œì¸í‰ê· : {person_averages.max():.2f}")
        print(f"ê°€ì¥ ë‚®ì€ ê°œì¸í‰ê· : {person_averages.min():.2f}")
        print(f"ê°œì¸í‰ê· ì˜ í‘œì¤€í¸ì°¨: {person_averages.std():.2f}")
        
        # 6. ìµœê³ ì , ìµœì €ì  í•™ìƒ ì°¾ê¸°
        print("\n" + "="*50)
        print("ğŸ† ì„±ì  ìˆœìœ„:")
        print("-" * 20)
        
        best_student_idx = person_averages.idxmax()
        worst_student_idx = person_averages.idxmin()
        
        print(f"ìµœê³  í‰ê· : {df.loc[best_student_idx, name_col]} ({person_averages[best_student_idx]:.2f})")
        print(f"ìµœì € í‰ê· : {df.loc[worst_student_idx, name_col]} ({person_averages[worst_student_idx]:.2f})")
        
        # 7. ê° ê³¼ëª©ë³„ 1ë“± í•™ìƒ
        print("\nğŸ“š ê³¼ëª©ë³„ 1ë“±:")
        print("-" * 20)
        for col in numeric_cols:
            top_student_idx = df[col].idxmax()
            top_score = df.loc[top_student_idx, col]
            top_student_name = df.loc[top_student_idx, name_col]
            print(f"{col}: {top_student_name} ({top_score}ì )")
        
        # 8. ì „ì²´ í†µê³„ ìš”ì•½
        print("\n" + "="*50)
        print("ğŸ“Š ì „ì²´ í†µê³„ ìš”ì•½:")
        print("-" * 50)
        print(df[numeric_cols].describe())
        
    else:
        print("\nâš ï¸  ìˆ«ìí˜• ë°ì´í„°ê°€ ì—†ì–´ì„œ í†µê³„ ê³„ì‚°ì„ í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        print("ë°ì´í„° íƒ€ì… ì •ë³´:")
        print(df.dtypes)
        
else:
    print("íŒŒì¼ì„ ì½ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. íŒŒì¼ ê²½ë¡œì™€ í˜•ì‹ì„ í™•ì¸í•´ì£¼ì„¸ìš”.")
```
